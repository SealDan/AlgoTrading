   - **Question**: You have a list of stock prices over 30 days. Write a logical expression that will identify consecutive price drops over three days or more. For instance, `[120, 118, 115, 110]` would qualify, but `[120, 118, 120, 110]` would not.

```python
#Write your code here
price = [120, 118, 115, 110]
days = 0
ans = False
temp = price[0]

for i in price:
    if(temp > i):
        days += 1
        if(days >= 3):
            ans = True
            break
    else:
        days = 0
    temp = i

print("The following list of stock prices is identified to be " + str(ans) + " to have consecutive price drops over three days or more!")

    
```

9. **Optimization with Loops**  
   - **Question**: Calculate the best day to buy and the best day to sell to maximize profit within a list of daily prices using a single pass of nested loops.

```python
#Write your code here
prices = [
    100, 101, 99, 102, 103, 104, 105, 106, 108, 107,
    106, 108, 110, 111, 113, 115, 114, 117, 118, 120,
    122, 121, 123, 125, 124, 126, 127, 129, 131, 133,
    134, 136, 137, 139, 138, 140, 142, 141, 143, 145,
    144, 147, 149, 148, 150, 151, 153, 154, 155, 156
]

best_profit = 0

for i in range(len(prices) - 1):
    buy_price = prices[i]
    
    for j in range(i+1, len(prices)):
        sell_price = prices[j]
        profit = sell_price - buy_price

        if profit > best_profit:
            best_profit = profit
            best_buy_day = i
            best_sell_day = j


print("Best Profit:", best_profit)
print("Best Buy Day:", best_buy_day)
print("Best Sell Day:", best_sell_day)
```

### Python Mini Project 2 :

Scenario: Simulated Stock Trading Bot

**Objective:**
Create a Python script that simulates trading over a dataset containing historical stock prices. The bot will decide whether to buy, hold, or sell based on the momentum calculated as the difference between the current price and the average price of the last few days.


### Project Components:

1. **Data Structure**:
   - Use a list of tuples or dictionaries to simulate historical stock data with each element containing information about the stock price on a particular day.

2. **Momentum Calculation**:
   - Calculate the average price of the last N days and compare it with the current day's price to determine the momentum.

3. **Trading Decisions**:
   - Use if-else statements to make trading decisions based on the calculated momentum.

4. **Trading Simulation**:
   - Loop through the dataset and simulate trading decisions, updating the portfolio accordingly.

5. **Performance Evaluation**:
   - At the end of the simulation, calculate and print the final portfolio value and compare it with the initial value to gauge performance.

### Skills and Python Features Used:
- **Python Operators**: Use arithmetic operators to calculate averages, gains, and losses.
- **Python if-else**: Make conditional decisions based on momentum and other trading indicators.
- **Python Modules**: Use the `datetime` module for handling dates, `matplotlib` for plotting stock prices and portfolio value over time.
- **Python While Loop**: Could be used to allow the simulation to run until a certain condition is met, like a stop-loss limit.
- **Python for loop**: Iterate over each day in the stock dataset to simulate trading.




```python
#Write your code here
import datetime as dt

prices = [
    100, 102, 101, 105, 107,
    106, 108, 110, 109, 112,
    115, 114, 118, 120, 119,
    121, 123, 125, 124, 127,
    130, 129, 131, 133, 132,
    135, 137, 139, 138, 140
]

start_date = dt.date(2025, 1, 1)
data = []

for i, p in enumerate(prices):
    data.append({
        "date": start_date + dt.timedelta(days=i),
        "price": p
    })

#parameters
N = 5
threshold = 1.0
initial_cash = 10_000.00

#portfolio
cash = initial_cash
shares = 0
portfolio_values = [] #date, portfolio value
signals = [] #date, buy/sell/hold
prices_over_time = []
```

```python
#calculate rolling average of N days
def average_last_n(data_list, i, N):
    window = data_list[i-N : i]
    total =  sum(d["price"] for d in window)
    return total/N
```

```python
for i in range(len(data)):
    date = data[i]["date"]
    price = data[i]["price"]
    day = data[i]

    #not able to calculate average over N first days
    if i >= N:
        avg_price = average_last_n(data, i, N)
        momentum = price - avg_price

        #trading decision
        if momentum > threshold:
            if cash > 0:
                shares_to_buy = cash // price
                if shares_to_buy > 0:
                    cash -= shares_to_buy*price
                    shares += shares_to_buy
                    signal = "BUY"
                else:
                    signal = "HOLD"
            else:
                signal = "HOLD"
    
        elif momentum < -threshold:
            if shares > 0:
                cash += shares*price
                shares = 0
                signal = "SELL"
            else:
                signal = "HOLD"
        else:
            signal = "HOLD"
            
    else:
        avg_price = None
        momentum = None
        signal = "HOLD"

    portfolio_value = cash + shares * price

    portfolio_values.append((date, portfolio_value))
    prices_over_time.append((date, price))
    signals.append((date, signal))

    print(f"{date} | Price: {price} | Signal: {signal} | Cash: {cash:.2f} | Shares: {shares} | Value: {portfolio_value:.2f}")
    
        
```

```python
import matplotlib.pyplot as plt
final_value = portfolio_values[-1][1]
profit = final_value - initial_cash
print("\n=== Performance Summary ===")
print(f"Initial Portfolio Value: {initial_cash:.2f}")
print(f"Final   Portfolio Value: {final_value:.2f}")
print(f"Net Profit/Loss        : {profit:.2f}")


# 6. Plot stock price and portfolio value over time
dates = [d for d, _ in prices_over_time]
price_series = [p for _, p in prices_over_time]
value_series = [v for _, v in portfolio_values]

plt.figure()
plt.plot(dates, price_series, label="Stock Price")
plt.plot(dates, value_series, label="Portfolio Value")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Stock Price vs Portfolio Value")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
